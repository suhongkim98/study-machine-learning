#### Index 연산

```
idx1 = pd.Index([1,2,4,6,8])
idx2 = pd.Index([2,4,5,6,7])

print(idx1.append(idx2))
print(idx1.difference(idx2))
print(idx1.intersection(idx2), idx1 & idx2) # 공통부분
print(idx1.union(idx2), idx1 | idx2) # 합집합, 중복을 빼고 append
print(idx1.delete(0)) # 0번 인덱스 삭제
print(idx1.drop(1)) # 1이라는 값 삭제
print(idx1 ^ idx2) # 여집합
```

### DataFrame 인덱싱

|사용 방법|설명|
|------|---|
|df[val]|하나의 컬럼 또는 여러 컬럼을 선택|
|df.loc[val]|라벨값으로 row의 부분집합 선택|
|df.loc[:, val]|라벨값으로 column의 부분집합 선택|
|df.loc[val1, val2]|라벨값으로 row와 col의 부분집합 선택|
|df.iloc[where]|정수 색인으로 row의 부분집합 선택|
|df.iloc[:, where]|정수 색인으로 col의 부분집합 선택|
|df.iloc[where_i, where_j]|정수 색인으로 row와 col의 부분집합 선택|
|df.at[label_i, label_j]|로우와 컬럼의 라벨로 단일값 선택|
|df.iat[i, j]|row와 col의 정수 색인으로 단일 값 선택|
|reindex|하나 이상의 축을 새로운 색인으로 재색인|
|get_value, set_value|로우와 컬럼의 이름으로 값 선택|

### 고성능 연산
* pd.eval(내용)
* df.eval(내용) 처럼 데이터프레임 접근한 상태로 고성능연산 가능

```
# pandas 자체가 연산을 빠르게 해주지만 더 빠르게 하기위해 고성능 연산을 하기도 한다
nrows, ncols = 100000, 100
df1, df2, df3, df4 = (pd.DataFrame(np.random.rand(nrows, ncols)) for i in range(4))
```

```
%timeit df1 + df2 + df3 + df4
```

```
%timeit pd.eval('df1 + df2 +df3 +df4') # 고성능연산
```

```
%timeit df1 * -df2 / (-df3 * df4)
```

```
%timeit pd.eval('df1 * -df2 / (-df3 * df4)')
```

```
df = pd.DataFrame(np.random.rand(1000000, 5), columns=['A','B','C','D','E'])
df.head()
```

```
%timeit df.eval('A + B / C - D * E')
```

```
df.eval('R = A + B / C - D * E', inplace=True) # 결과를 컬럼에 추가할 수 있다
df.head()
```

```
col_mean = df.mean(1)
df['A'] + col_mean
```

|---|---|

### 리샘플링(Resampling)

* 리샘플링(Resampling): 시계열의 빈도(freq) 변환
* 다운샘플링(Down sampling): 상위 빈도 데이터를 하위 빈도 데이터로 집계
* 업샘플링(Up sampling): 하위 빈도 데이터를 상위 빈도 데이터로 집계

|인자|설명|
|---|---|
|freq|리샘플링 빈도|
|axis|리샘플링 축(기본 axis=0)|
|fill_method|업샘플링시 보간 수행(None, ffill, bfill)|
|closed|다운샘플링 시 각 간격의 포함 위치(right, left)|
|label|다운샘플링 시 집계된 결과 라벨 결정(right, left)|
|loffset|나뉜 그룹의 라벨을 맞추기 위한 오프셋|
|limit|보간법을 사용할 때 보간을 적용할 최대 기간|
|kind|기간(period) 또는 타임스탬프 집계 구분|
|convention|기간을 리샘플링할 때 하위 빈도 기간에서 상위 빈도로 변환시 방식(start 또는 end)|

```
dr = pd.date_range('2020-01-01', periods=200, freq='D')
ts = pd.Series(np.random.randn(len(dr)), index=dr)
ts
```

### 무빙 윈도우(Moving Window)

```
df = pd.DataFrame(np.random.randn(300,4),
                  index=pd.date_range('2020-01-01',periods=300,freq='D'),
                  columns=['C1','C2','C3','C4'])
df
```

```
# 테이블로 보기 불편하면 .plot() 붙이면 그림으로 바꿔줌
df.rolling(30).mean().plot() # 30일치 이동평균 구하기
```

```
df.rolling(60).mean().plot()
```

```
df.C1.rolling(60, min_periods=10).std().plot() # 최소 10일 간격을 주고 60일치 이동평균
```

```
df.rolling(60, min_periods=10).std()[10:50].plot() # 슬라이싱
```

```
df.rolling(60, min_periods=10).std().expanding().mean().plot() # expanding()쓰면 확장 윈도우를 쓰게됨
```

```
df.rolling(60).mean().plot(logy=True) # 로그스케일로
```

```
df.rolling('20D').mean().plot()
```

```
df.C1.rolling(30, min_periods=20).mean().plot(style='--', label='Simple MA')
df.C1.ewm(span=30).mean().plot(style='-', label='EWMA') # 롤링, 익스펜딩 같이 쓸 떄 ewm을 씀
```

```
df.C1.rolling(100, min_periods=50).corr(df.C3).plot() # .corr붙여서 상관분석 가능
```

```
df.C2.rolling(100, min_periods=50).corr(df.C4).plot() # .corr붙여서 상관분석 가능
```
